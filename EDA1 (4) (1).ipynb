```python
#Load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data_clean.csv")
data
```

```python
np.mean(data["Ozone"])
```

```python
data.info()
```

```python
# Dataframe attributes
print(type(data))
print(data.shape)
print(data.size)
```

```python
#Drop duplicate column(Temp c) and unnamed column
data1 = data.drop(['Unnamed: 0',"Temp C"], axis = 1)
data1
```

```python
data1.info()
```

```python
data1.info()
```

```python
data2 = data.drop(['Unnamed: 0',"Temp C","Ozone","Solar.R","Wind","Month","Day","Year","Temp","Weather"], axis =1)
data2
```

```python
data1['Month']=pd.to_numeric(data['Month'],errors='coerce')
data1.info()
```

```python
data1[data1.duplicated(keep = False)]
```

```python
data1[data1.duplicated()]
```

```python
#Drop duplicated rows
data1.drop_duplicates(keep='first', inplace = True)
data1
```

```python
# change column names (Rename the columns)
data1.rename({'Solar.R': 'Solar'}, axis=1, inplace = True)
data1
```

```python
#display data1 info()
data1.info()
```

```python
#Display data1 missing values count in each column using isnull().sum
data1.isnull().sum()
```

```python
#Visualize data1 missing values using graph

cols = data1.columns
colors = ['red', 'green']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar = True)
```

```python
cols = data1.columns
colors = ['black', 'green']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar = True)
```

```python
#Imputation of mising values with median
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean()
print("Median of Ozone: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
#Replace the ozone missing values with median values
data1["Ozone"] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
#Imputation of mising values with median
mean_solar = data1["Solar"].mean()
mean_solar = data1["Solar"].mean()
print("Mean of Solar: ", mean_solar)
print("Mean of Solar: ", mean_solar)
```

```python
#Replace the Solar missing values with mean value
data1["Solar"] = data1['Solar'].fillna(mean_solar)
data1.isnull().sum()
```

```python
data1.head()
```

```python
print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python
data1["Weather"] = data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
mode_month = data1["Month"].mode()[0]
data1["Month"] = data1["Month"].fillna(mode_month)
data1.isnull().sum()
```

```python
#Reset the index column
data1.reset_index(drop=True)
```

```python
#create a figure with two subplots, stacked vertically
fig, axes = plt.subplots(2,1,figsize=(8,6),gridspec_kw={'height_ratios': [1,3]})

#plot the boxplot in the first (top) subplot
sns.boxplot(data=data1["Ozone"], ax=axes[0], color='skyblue', width=0.5, orient = 'h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Ozone Levels")

# Plot the histogram with KDE curve in the second (bottom) subplot
sns.histplot(data1["Ozone"], kde=True, ax=axes[1], color='purple', bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Ozone Levels")
axes[1].set_ylabel("Frequency")

# Adjust loyout for better spacing
plt.tight_layout()

plt.show()
```

```python
fig,axes=plt.subplots(2,1,figsize=(8,6),gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["Ozone"],ax=axes[0],color='skyblue',width=0.5,orient='h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Ozone Levels")

sns.histplot(data1["Ozone"],kde=True,ax=axes[1],color='Purple',bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Ozone Levels")
axes[1].set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
mu = data1["Ozone"].describe()[1]
sigma = data1["Ozone"].describe()[2]

for x in data1["Ozone"]:
    if ((x < (mu - 3*sigma)) or (x > (mu + 3*sigma))):
        print(x)
```

```python
#Quantile-Quantile plot for detection of outliers
import scipy.stats as stats

#create Q-Q plot
plt.figure(figsize=(8,6))
stats.probplot(data1["Ozone"], dist="norm", plot=plt)
plt.title("Q-Q Plot for Quantlier Detection", fontsize=14)
plt.xlabel("Theoretical Quantiles", fontsize=12)
```

```python
#Observation from Q-Q plot
 >The data does not follow normal distribution as the data points are deviating significantly away from the red line
 >The data shows a right-skewed distribution and possible  outliers
```
