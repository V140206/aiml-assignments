```python

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# Generating random data
data1 = {
    'Age': np.random.randint(18, 70, size=20),
    'salary': np.random.randint(30000, 120000, size=20),
    'Purchased': np.random.choice([0, 1], size=20),
    'Gender': np.random.choice(['Male', 'Female'], size=20),
    'City': np.random.choice(['Hyderabad', 'Goa', 'America'], size=20)
}
df = pd.DataFrame(data1)

df.loc[3, 'Age'] = np.nan
df.loc[7, 'salary'] = np.nan

print("Data before mean cleaning:")
print(df)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['salary'].fillna(df['salary'].mean(), inplace=True)

print("\nData after mean cleaning:")
print(df)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

def generate_example_dataset():
    data = {
        'PassengerId': [1, 2, 3, 4, 5],
        'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
        'Age': [22, None, 25, None, 30],
        'Fare': [7.25, 71.28, None, 8.05, 10.5],
        'Survived': [0, 1, 1, 0, 1]
    }
    return pd.DataFrame(data)

def clean_dataset(df):
    print("Original Dataset:")
    print(df)
    numeric_df = df.select_dtypes(include='number')

    imputer = SimpleImputer(strategy='mean')
    
    # Apply imputation and store the cleaned numeric data
    df_cleaned_numeric = pd.DataFrame(imputer.fit_transform(numeric_df), columns=numeric_df.columns)
    
    # Combine cleaned numeric data back into the original DataFrame
    df_cleaned = df.copy()
    df_cleaned[numeric_df.columns] = df_cleaned_numeric
    
    print("\nCleaned Dataset:")
    print(df_cleaned)

# Corrected entry point
if __name__ == "_main_":
    example_dataset = generate_example_dataset()
    clean_dataset(example_dataset)
```
